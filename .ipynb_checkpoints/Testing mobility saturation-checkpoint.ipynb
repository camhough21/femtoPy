```python
from femtoPy.imports import *
import femtoPy.diffusion as dif

%matplotlib inline
```

```python
N=np.logspace(14,18,100)
N0=1.05e17
tauMax=375
tauMin=5
alpha=0.22

tau=dif.Caughey_Thomas(N,N0,tauMax,tauMin,alpha)

fig,ax=figure()
ax.scatter(N,tau)
ax.set_xscale('log')

ax.set_ylabel(r'$\tau$ (fs)')
ax.set_xlabel(r'$\rho$ ($\times$10$^{18}$ cm$^{-3}$)')


mu=dif.mobility(tau=tau*10**-15)
muEff=dif.ambipolar_mobility(mu_e=mu,mu_h=0.0415)
fig,ax=figure()
# ax.scatter(N,mu,label='electron')
ax.scatter(N,muEff,label='ambipolar')
ax.set_xscale('log')
ax.legend()

ax.set_ylabel(r'$\mu$ ($m^2V^{-1}s^{-1}$)')
ax.set_xlabel(r'$\rho$ ($\times$10$^{18}$ cm$^{-3}$)')

D=dif.diffusion_coefficient(muEff,T=300)
fig,ax=figure()
ax.scatter(N,D)
ax.set_xscale('log')

ax.set_ylabel(r'D (??)')
ax.set_xlabel(r'$\rho$ ($\times$10$^{18}$ cm$^{-3}$)')

ax.set_ylim(0.0007,0.0011)
```
